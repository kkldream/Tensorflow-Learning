```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
test_dir = "dataset/test"
```

```python
def show_img_matrix(generator):
    output = np.zeros((32 * 8, 32 * 8, 3), dtype=np.uint8)
    for col in range(8):
        for row in range(8):
            img = generator[col][0][row]
            output[col*32:(col+1)*32, row*32:(row+1)*32] = img
    plt.figure(figsize=(8, 8))
    plt.imshow(output)
```

```python
datagen = ImageDataGenerator()
generator = datagen.flow_from_directory(test_dir, shuffle=False, batch_size=8, class_mode="categorical", target_size=(32, 32))
show_img_matrix(generator)
```

```python
datagen = ImageDataGenerator(rotation_range=360,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             vertical_flip=True)
generator = datagen.flow_from_directory(test_dir, shuffle=False, batch_size=8, class_mode="categorical", target_size=(32, 32))
show_img_matrix(generator)
```

```python
datagen = ImageDataGenerator(rotation_range=360,
                             shear_range=0.4,
                             zoom_range=0.2,
                             width_shift_range=0.3,
                             height_shift_range=0.3,
                             horizontal_flip=True,
                             vertical_flip=True)
generator = datagen.flow_from_directory(test_dir, shuffle=False, batch_size=8, class_mode="categorical", target_size=(32, 32))
show_img_matrix(generator)
```
